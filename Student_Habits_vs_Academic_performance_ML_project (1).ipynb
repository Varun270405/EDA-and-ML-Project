# Project Summary

This project focuses on analyzing the factors that influence students' academic performance using Exploratory Data Analysis (EDA) and Linear Regression. The dataset consists of 1,000 student records with information related to demographic details, study habits, lifestyle choices, attendance, health, and examination scores. The primary objective of the project was to identify the key variables affecting students' exam performance and develop a predictive model capable of estimating exam scores based on these factors.

The project began with data preprocessing, where the dataset was inspected for missing values, duplicate records, and data inconsistencies. A small number of missing values were found in the parental education level column and were handled using mode imputation to preserve the dataset's integrity. The student ID column was removed since it did not contribute to the prediction process. Descriptive statistics were generated to understand the distribution, central tendency, and variability of numerical variables, while categorical variables were examined using frequency distributions.

Exploratory Data Analysis was then performed to uncover patterns and relationships within the data. Various visualizations, including histograms, box plots, count plots, scatter plots, bar charts, and a correlation heatmap, were created to analyze the characteristics of each variable. The analysis showed that students who dedicated more time to studying and maintained higher attendance percentages generally achieved better examination scores. Healthy lifestyle habits, such as adequate sleep and regular physical exercise, also demonstrated a positive association with academic performance. Conversely, excessive time spent on social media and streaming platforms appeared to have a negative impact on students' scores. The visualizations also highlighted the influence of factors such as diet quality, internet availability, mental health, and extracurricular participation on overall academic outcomes.

Following EDA, the categorical variables were transformed into numerical format using one-hot encoding to prepare the dataset for machine learning. The processed data was divided into training and testing sets, ensuring that the model could be evaluated on unseen data. A Linear Regression model was then trained using the selected independent variables to predict the target variable, exam score. Model performance was evaluated using standard regression metrics, including Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and the R² Score. These metrics provided valuable insights into the prediction accuracy and the model's ability to explain variations in student performance.

The regression coefficients further revealed the contribution of each feature toward the predicted exam score. Variables such as study hours per day and attendance percentage positively influenced academic performance, whereas increased social media usage and Netflix viewing hours generally reduced predicted scores. Lifestyle and well-being factors, including sleep duration, exercise frequency, and mental health ratings, also contributed meaningfully to the prediction model.

Overall, this project demonstrates the practical application of data analysis and machine learning techniques in the field of education. By combining data preprocessing, visualization, statistical analysis, and predictive modeling, the study successfully identified the major determinants of student performance and provided meaningful insights that can support students, educators, and academic institutions in making informed decisions. The findings suggest that encouraging consistent study habits, maintaining regular attendance, promoting healthy lifestyles, and minimizing excessive digital distractions can significantly improve academic outcomes. Furthermore, the developed regression model serves as a useful tool for predicting student performance and can assist educational institutions in identifying students who may require additional academic guidance or support.

#### 3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

1. Mean Absolute Error (MAE)

* Business Impact: Helps educational institutions identify students needing academic support more accurately.

2. Mean Squared Error (MSE)

* Business Indication: Measures the average squared prediction error, giving more weight to large errors. Lower MSE indicates fewer major prediction mistakes.
* Business Impact: Improves the reliability of student performance predictions for better academic planning.

3. Root Mean Squared Error (RMSE)

* Business Indication: Represents the average prediction error in the same unit as exam scores. Lower RMSE means better model performance.
* Business Impact: Enables more dependable predictions for student performance and intervention strategies.

4. Mean Absolute Percentage Error (MAPE)

* Business Indication: Measures the average prediction error as a percentage of the actual exam scores. Lower MAPE indicates higher prediction accuracy in percentage terms.
* Business Impact: Helps institutions understand the model's prediction error in an intuitive percentage format, making it easier to evaluate and compare model performance.

5. R² Score (Coefficient of Determination)

* Business Indication: Indicates how well the model explains the variation in exam scores. A higher R² Score means the model fits the data better.
* Business Impact: Helps identify the key factors influencing student performance, enabling data-driven decisions to improve academic outcomes.
